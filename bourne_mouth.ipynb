```python
bourne=sales[['Venue Category','Venue Latitude','Venue Longitude']]
bourne.head()
```

```python
category_group=bourne.groupby(['Venue Category']).sum()
category_group
```

```python
category_group.unstack()
```

```python

my_plot = category_group.unstack().plot(kind='bar',
                             title="Total")
my_plot.set_xlabel("Venue Longitude")
my_plot.set_ylabel("Venue Latitude")
```

```python
plt.show()
```

```python
my_plot = category_group.unstack().plot(kind='bar',stacked=True,
                                    title="Total",
                                        figsize=(9, 7))
my_plot.set_xlabel("Customers")
my_plot.set_ylabel("Sales")
my_plot.legend(["Total","Coffee shop","Park","Ice Cream Shop"], loc=9,ncol=4)
```
